```python
import sqlite3

connection = sqlite3.connect('books.db')

import pandas as pd

pd.options.display.max_columns = 10

pd.read_sql('SELECT * FROM authors', connection, index_col=['id'])
```

```python
pd.read_sql('SELECT * FROM titles', connection)
```

```python
pd.read_sql('SELECT * FROM author_ISBN', connection)
```

```python
pd.read_sql("""SELECT title, edition, copyright FROM TITLES where copyright > '2017'""", connection)
```

```python
pd.read_sql("""SELECT id, first, last FROM authors WHERE last LIKE 'D%'""",connection)
```
